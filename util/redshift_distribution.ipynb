```python
# Libraries to use

from astropy.io import fits
import pandas as pd
import matplotlib.pyplot as plt
```

Here is the code to get the redshift distribution of diferrent simulations of a specific object:

```python
# Load FITS files, you can change it for wich one you want
light_curve = fits.open("../AGN_20241121_12of50/AGN_20241121_12of50_PHOT.FITS")[1].data
head = fits.open("../AGN_20241121_12of50/AGN_20241121_12of50_HEAD.FITS")[1].data

# Histogram

plt.hist(head['SIM_REDSHIFT_CMB'],bins=30, edgecolor="black")

# Etiquetas y título
plt.xlabel('Redshifts')
plt.ylabel('Frecuency')
plt.title('Redshift  distribution')

# Mostrar la gráfica
plt.show()
```
